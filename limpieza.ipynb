```python
def data_exploring(csv):
    forma = csv.shape
    print(f"La forma es {forma}")
    print("_______________")
    columnas = csv.columns
    print(f"Las columnas son {columnas}")
    print("_______________")
    nulos = csv.isna().sum().reset_index()
    print(f"Los nulos son:")
    display(nulos)
    print("_______________")
    duplicados = csv.duplicated().sum()
    print(f"Hay {duplicados} duplicados")
    print("_______________")
    tipo_dato = csv.dtypes.reset_index()
    print(f"Los datos son de tipo:")
    display(tipo_dato)
```

```python
def limpiar(dataframe):
    for col in dataframe.columns:
        # Verificar si el nombre de la columna no es 'track_name' ni 'artist(s)_name'
        if col not in ['trackname', 'artist(s)name']:
            # Verificar si el tipo de datos de la columna es una cadena
            if dataframe[col].dtype == 'object':
                # Reemplazar la coma (,) con un punto (.)
                dataframe[col] = dataframe[col].str.replace(",", "")
                # Intentar convertir los valores a números enteros si es posible
                try:
                    dataframe[col] = dataframe[col].astype(int)
                except ValueError:
                    pass  # Ignorar si no se puede convertir a int
```
